```python
%%sql

-- Query 1: Calculate the average salary for each department and show departments with an average salary above 55,000
-- Assuming there is a 'Departments' table with 'DepartmentID', 'DepartmentName', and 'Salary' columns
-- Modify the table and column names as per your actual schema
-- This query will need to be modified or omitted if you do not have a Departments table
SELECT Product_Type, AVG(Price) AS Average_Price
FROM Product_Data
GROUP BY Product_Type
HAVING AVG(Price) > 2;
```

```python
%%sql

-- Query 3: Join the Orders table with the Product_Data table based on the Product_Name column, showing the product types instead of product names in the Orders table
SELECT o.OrderID, o.Quantity, p.Product_Type, o.Price, o.Order_Date, o.Ship_Date, o.Discount, o.Total_Amount
FROM Orders o
INNER JOIN Product_Data p ON o.Product_Name = p.Product_Name;
```

```python
%%sql

-- Query 4: Calculate the average total amount for each product type and filter out only those product types where the average total amount is greater than 5
-- Using Inner Join, Group By and Having
SELECT p.Product_Type, AVG(o.Total_Amount) AS Average_Total_Amount
FROM Orders o
INNER JOIN Product_Data p ON o.Product_Name = p.Product_Name
GROUP BY p.Product_Type
HAVING AVG(o.Total_Amount) > 5;
```

```python
%%sql

-- Query 5: Retrieve a list of orders along with their product details using Inner Join
SELECT o.OrderID, o.Product_Name, p.Price, p.Product_Type, o.Quantity, o.Order_Date, o.Ship_Date, o.Discount, o.Total_Amount
FROM Orders o
INNER JOIN Product_Data p ON o.Product_Name = p.Product_Name;
```

```python
%%sql

-- Query 6: Retrieve all products along with the number of orders made for each product, including products that have not been ordered, using Left Join
SELECT p.Product_Name, p.Price, p.Product_Type, COUNT(o.OrderID) AS Order_Count
FROM Product_Data p
LEFT JOIN Orders o ON p.Product_Name = o.Product_Name
GROUP BY p.Product_Name;
```

```python
%%sql

-- Query 7: Retrieve all orders along with their product details, including products that do not have any orders, using Right Join
-- Note: SQLite does not support RIGHT JOIN directly, we can achieve this by switching the table order and using LEFT JOIN
SELECT o.OrderID, o.Product_Name, p.Price, p.Product_Type, o.Quantity, o.Order_Date, o.Ship_Date, o.Discount, o.Total_Amount
FROM Product_Data p
LEFT JOIN Orders o ON p.Product_Name = o.Product_Name;
```

```python
%%sql

-- Query 8: Retrieve a Cartesian product of all rows from both tables, using Cross Join
SELECT *
FROM Orders o
CROSS JOIN Product_Data p;
```
